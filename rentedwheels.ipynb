```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("rentedwheels_vehicle1.csv")
df
```

```python
features = ['ac', 'acceleration_time', 'auto_manual', 'file_path', 'fuel_electric','max_power','seats','top_speed','type','vehicle_image','vehicle_name','vehicle_number','vehicle_price','vehicle_rating','vehicle_review','vehicle_type','vehicle_status']
for feature in features:
    df[feature] = df[feature].fillna('')
```

```python
def combined_features(row):
    return str(row['ac']) + " " + str(row['acceleration_time']) + " " +str(row['auto_manual']) + " " + str(row['file_path']) + " " +str(row['fuel_electric']) + " " + str(row['max_power']) + " " + str(row['seats']) + " " + str(row['top_speed']) + " " + str(row['type']) + " " + str(row['vehicle_image']) + " " + str(row['vehicle_number']) + " " + str(row['vehicle_price']) +" " + str(row['vehicle_rating']) + " " + str(row['vehicle_review']) + " " + str(row['vehicle_type']) + " " + str(row['vehicle_status']) + " " + str(row['file_path'])
df["combined_features"] = df.apply(combined_features, axis =1)
```

```python
cv = CountVectorizer()
```

```python
count_matrix = cv.fit_transform(df["combined_features"])
```

```python
count_array = count_matrix.toarray()
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
vis = pd.DataFrame(data = cosine_sim)
```
